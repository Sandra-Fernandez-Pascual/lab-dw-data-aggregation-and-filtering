2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
yes_customers = df[df["Response"] == "Yes"]
```

```python
pivot = yes_customers.pivot_table(
    values=[
        "Monthly Premium Auto",
        "Customer Lifetime Value",
        "Total Claim Amount"
    ],
    index=["Policy Type", "Gender"],
    aggfunc="mean"
).round(2)

pivot
```

```python
#Among customers who responded **"Yes"**, **Personal Auto** customers show mixed results. 
#Female customers have the highest **Customer Lifetime Value (CLV), which is the estimated total value a customer brings to the 
#company over their entire relationship with it** (8339.79) and the highest monthly premium (99.00). However, both male and female Personal Auto 
#customers also have the highest average total claim amounts (452.97 and 457.01), meaning they generate high revenue but also higher costs.

#**Corporate Auto** customers appear to be the most balanced segment. Both male and female customers have relatively high CLVs and monthly premiums, 
#while also having the lowest average total claim amounts. In particular, **Male Corporate Auto** customers have the lowest average claim amount 
#(408.58), making this segment appear both profitable and low-risk.

#**Special Auto** customers fall between the other two policy types. Male customers have a relatively high CLV (8247.09), 
#while female customers have a lower CLV (7691.58). Their monthly premiums are slightly lower than those of the other policy types, 
#and their average claim amounts are moderate. This suggests that Special Auto customers are a reasonably profitable segment with a moderate 
#level of risk.
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_summary = df.groupby(["Education", "Gender"])["Customer Lifetime Value"].agg(
    ["max", "min", "median"]
).round(2)

clv_summary
```

```python
#The results show some differences in Customer Lifetime Value (CLV), which is the estimated total value a customer brings to the company over 
#their entire relationship with it, across education levels and gender.

#Customers with *High School or Below* education have the highest maximum CLV, especially males (83,325.38), suggesting that some customers 
#in this segment can generate very high value for the company. In contrast, *Doctor* degree holders have the lowest maximum CLV, particularly 
#males (32,677.34).

#Looking at the **median CLV**, which represents the middle value and is less affected by extremely high or low values, *High School or Below* 
#customers also have the highest median CLV for both females (6,039.55) and males (6,286.73). This suggests that this education group generally 
#provides higher customer value than the others, not just a few exceptional customers.

#Overall, the differences between males and females are relatively small within most education levels. 
#The largest pattern is that customers with *High School or Below* education tend to have the highest CLV, while customers with a 
#*Doctor* degree tend to have lower CLV.
```
